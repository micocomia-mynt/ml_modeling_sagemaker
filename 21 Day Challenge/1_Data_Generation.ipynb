This notebook generates a synthetic dataset with two classes and 1000 samples. To have an idea of how the classes are split, the synthetic dataset was visualized using scatter and pair plots. For the purposes of training a model, the generated were split into a 60-20-20 training-validation-test split. Finally, the split data sets were exported as CSV files for later use.

---

Steps:
- GENERATE 1000 (ONE THOUSAND) RECORDS USING THE SKLEARN LIBRARY
- PREPARE A QUICK DIAGRAM / CHART OF THE PROPERTIES OF THE DATASET USING MATPLOTLIB
- PERFORM THE TRAIN-TEST SPLIT (TRAIN-VALIDATION-TEST)
- STORE THE FILES INSIDE A DIRECTORY FOR LATER USE

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import sklearn.datasets as dt

from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
```


### I. Generating 1000 records
---

Usin sklearn's make_classification method, we create a synthetic dataset with 1000 samples and 5 features. Among these 5 features, 3 are informative and 2 are redundant. For the output, we have 2 classes (0 and 1).

```python
X,y = dt.make_classification(n_samples = 1000, n_features = 5, 
                             n_informative = 3, n_redundant = 2,
                             n_classes = 2, n_clusters_per_class = 2,
                             random_state = 43)
```

To validate the number of samples, we check the size of arrays X and Y:

```python
print(f"Size of X: {len(X)}")
print(f"Size of Y: {len(y)}")
```

### II. Diagram/chart of the properties
---

First, we create dataframes for our X and y arrays to facilitate the creation of diagrams.

```python
df_X = pd.DataFrame(X, columns = ['X1','X2','X3','X4','X5'])
df_y = pd.DataFrame(y, columns = ['y'])
```

Since the synthetic dataset is multi-dimensional, we cannot directly plot the features with respect to Y. To compress the feature dimensions into 2 so that we can plot the dataset in a 2D space, we employ Principal Component Analysis (PCA). The following fits our input, df_X to PCA and we specifiy the needed dimension in n_components which is 2.

```python
pca = PCA(n_components=2).fit(df_X)
pca_2d = pca.transform(df_X)
```

Finally, we match the values of our df_y with the compressed features and create a scatter plot. Features 1 and 2 serve as the X and Y of our scatter plot. The values of df_y serve as a guide on how to mark and color the different markers.

```python
for i in range(0, pca_2d.shape[0]):
	if df_y['y'].iloc[i] == 0:
		c1 = plt.scatter(pca_2d[i,0],pca_2d[i,1],c='r',    marker='+')
	elif df_y['y'].iloc[i] == 1:
		c2 = plt.scatter(pca_2d[i,0],pca_2d[i,1],c='g',    marker='*')
    
plt.legend([c1, c2], ['0', '1'])
plt.title('Training Dataset')
```

Meanwhile, the following pair plot shows us the relationships of the different features and the two classes per two features. We first concatenated our X and y dataframes, then specified the output as the hue attribute of the pair plot.

```python
df_Xy = pd.concat([df_X, df_y], axis = 1)
sns.pairplot(df_Xy, vars=df_Xy.columns[:-1], hue="y")
```

### III. Train-Validation-Test split
---

We create a train-validation-test split yung the train_test_split method from Sklearn. To ensure reproducibility, we set a value for the random_state attribute. The split is done twice, with the first one being for the 80-20 Train-Test split. To create the validation set, we further split the Training set. The end result is a 60-20-20 Train-Validation-Test split.

```python
# Splitting into 80-20 Train-Test splits.
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size = 0.2,
                                                    random_state = 43)
```

```python
# Splitting into 75-25 Train-Validation split or 60-20-20 Train-Validation-Test split.
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, 
                                                  test_size = 0.25,
                                                  random_state = 43)
```

Below, we can see the sizes of each set in a dataframe.

```python
sizes = [[len(X_train), len(y_train)],
         [len(X_val), len(y_val)],
         [len(X_test), len(y_test)]]
```

```python
df_sizes = pd.DataFrame(sizes, index = ['Training','Validation','Test'], columns = ['X','Y'])
df_sizes
```

### IV. Storing files
---

To store the sets as CSV files, we first convert the arrays into dataframes with their respective labels.

```python
df_X_train = pd.DataFrame(X_train, columns = ['X1', 'X2', 'X3','X4','X5'])
df_X_val = pd.DataFrame(X_val, columns = ['X1', 'X2', 'X3','X4','X5'])
df_X_test= pd.DataFrame(X_test, columns = ['X1', 'X2', 'X3','X4','X5'])

df_y_train = pd.DataFrame(y_train, columns = ['Y'])
df_y_val = pd.DataFrame(y_val, columns = ['Y'])
df_y_test = pd.DataFrame(y_test, columns = ['Y'])
```

Finally, we use Pandas' .to_csv method to store the converted DataFrames in our local machine.

```python
# Saving as CSV files
df_X_train.to_csv("data/X_train.csv", index = False)
df_X_val.to_csv("data/X_val.csv", index = False)
df_X_test.to_csv("data/X_test.csv", index = False)

df_y_train.to_csv("data/y_train.csv", index = False)
df_y_val.to_csv("data/y_val.csv", index = False)
df_y_test.to_csv("data/y_test.csv", index = False)
```
