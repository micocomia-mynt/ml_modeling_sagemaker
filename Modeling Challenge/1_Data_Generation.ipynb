```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets as dt

from sklearn.model_selection import train_test_split
```


### I. Generating 1000 records
---

```python
X,y = dt.make_classification(n_samples = 1000, n_features = 5, 
                             n_informative = 3, n_redundant = 2,
                            n_classes = 2, n_clusters_per_class = 2,
                            random_state = 43)
```

```python
print(f"Size of X: {len(X)}")
print(f"Size of Y: {len(y)}")
```

### II. Diagram/chart of the properties
---

```python
plt.scatter(X[:,0], X[:,1], c=y)
```

### III. Train-Validation-Test split
---

```python
# 80-20 Train-Test
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size = 0.2,
                                                    random_state = 43)
```

```python
# 75-25 Train-Validation (60-20-20 Train-Validation-Test split)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, 
                                                  test_size = 0.25,
                                                  random_state = 43)
```

```python
print(f"Size of X_train: {len(X_train)}")
print(f"Size of y_train: {len(y_train)}")
print(f"Size of X_val: {len(X_val)}")
print(f"Size of y_val: {len(y_val)}")
print(f"Size of X_test: {len(X_test)}")
print(f"Size of y_test: {len(y_test)}")
```

### IV. Storing files
---

```python
df_X_train = pd.DataFrame(X_train, columns = ['X1', 'X2', 'X3','X4','X5'])
df_X_val = pd.DataFrame(X_val, columns = ['X1', 'X2', 'X3','X4','X5'])
df_X_test= pd.DataFrame(X_test, columns = ['X1', 'X2', 'X3','X4','X5'])
df_y_train = pd.DataFrame(y_train, columns = ['Y'])
df_y_val = pd.DataFrame(y_val, columns = ['Y'])
df_y_test = pd.DataFrame(y_test, columns = ['Y'])
```
